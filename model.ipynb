```python
import matplotlib.pyplot as plt
import pandas as pd

# Example DataFrame (replace with your actual DataFrame)
data = {
    "EthnicGroup": ["GroupA", "GroupB", "GroupC", "GroupD", "GroupE", "GroupA", "GroupB", "GroupC", "GroupD", "GroupA"],
    # Add more rows as needed
}
df = pd.DataFrame(data)

# Calculate counts for each ethnic group
group_counts = {
    "GroupA": df[df['EthnicGroup'] == "GroupA"].shape[0],
    "GroupB": df[df['EthnicGroup'] == "GroupB"].shape[0],
    "GroupC": df[df['EthnicGroup'] == "GroupC"].shape[0],
    "GroupD": df[df['EthnicGroup'] == "GroupD"].shape[0],
    "GroupE": df[df['EthnicGroup'] == "GroupE"].shape[0]
}

# Create lists for labels and counts
labels = list(group_counts.keys())
counts = list(group_counts.values())

# Handle NaN values by replacing with 0 (if any)
counts = [count if count == count else 0 for count in counts]

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Ethnic Groups')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
